```python
import pandas as pd
import numpy as np
import glob

sessions = pd.DataFrame(columns = ['File', 'Peak', 'Mean', 'Mean2', 'Total']) 
for file in glob.glob("prj5pr4/*.dat"):
    if "throughput" in file and "prj5_p5_" in file:        
        df = pd.read_csv(file, sep=" ", header=None)
        values = df.iloc[:,1].values.tolist()
        # remove trailing zeros
        while values[-1] == 0:
            values.pop(-1)
            if len(values) == 1:
                break
        # remove zeros in beginning
        if len(values) > 1:
            while values[0] == 0:
                values.pop(0)
        values = np.array(values)
        total = values.sum()
        mean = values.mean()
        mean2 = (total) / (len(values)) if "64" in file else (total) / (len(values))
        peak = values.max()
        sessions = sessions.append({'File': file, 'Peak': peak, 'Mean': mean, 'Mean2': mean2, 'Total': total}, ignore_index=True)

sessions = sessions.sort_values(by ='File')
sessions.to_excel("analysis.xlsx")
sessions
```
